```python
import pandas as pd
import numpy as np

# --- sample data creation (agar tumhare paas sales.csv nahi hai, yeh bana kar test kar lo) ---
data = [
    [1, '2024-07-01', 'Punjab', 'Widget', 10, 120.0, 'Ali'],
    [2, '2024-07-05', 'Sindh',  'Gadget', 5,  300.0, 'Sara'],
    [3, '2024-07-08', 'Punjab', 'Widget', 3,  np.nan,  'Ali'],    # missing price
    [4, '2024-08-02', 'KPK',    'Gizmo',  7,  450.0, 'Rehan'],
    [5, '2024-08-15', 'Punjab', 'Gadget', 2,  300.0, 'Sara'],
    [6, '2024-09-01', 'Sindh',  'Widget', 20, 110.0, 'Zara'],
    [7, '2024-09-11', 'Punjab', 'Gizmo',  4,  500.0, 'Ali'],
    [8, '2024-09-20', 'KPK',    'Widget', 1,  115.0, 'Rehan'],
    [9, '2024-10-02', 'Sindh',  'Gizmo',  10, 480.0, 'Sara'],
    [10,'2024-10-05', 'Punjab', 'Gadget', 30, 290.0, 'Zara'],
]
df = pd.DataFrame(data, columns=['OrderID','Date','Region','Product','Units','UnitPrice','SalesRep'])

# agar real file hai to use karo:
# df = pd.read_csv('sales.csv', parse_dates=['Date'])

# 1) Date ko datetime
df['Date'] = pd.to_datetime(df['Date'])

# 2) Fill missing UnitPrice by median price per Product (task 7 before revenue calc)
df['UnitPrice'] = df.groupby('Product')['UnitPrice'].transform(
    lambda x: x.fillna(x.median())
)

# 3) Revenue column
df['Revenue'] = df['Units'] * df['UnitPrice']

# 4) Total revenue per Product (descending)
product_revenue = df.groupby('Product', as_index=False)['Revenue'].sum().sort_values('Revenue', ascending=False)
print("Total revenue per Product:\n", product_revenue, "\n")

# 5) Top 3 SalesRep by revenue
salesrep_top3 = df.groupby('SalesRep', as_index=False)['Revenue'].sum().sort_values('Revenue', ascending=False).head(3)
print("Top 3 SalesReps by revenue:\n", salesrep_top3, "\n")

# 6) Punjab monthly revenue trend
punjab = df[df['Region']=='Punjab'].copy()
punjab_monthly = (punjab.set_index('Date')
                         .resample('M')['Revenue']
                         .sum()
                         .reset_index()
                         .assign(Month=lambda x: x['Date'].dt.to_period('M').astype(str))
                         .loc[:, ['Month','Revenue']])
print("Punjab monthly revenue:\n", punjab_monthly, "\n")

# 7) Pivot table Units sold by Region x Product
pivot_units = df.pivot_table(index='Region', columns='Product', values='Units', aggfunc='sum', fill_value=0)
print("Pivot (Units by Region x Product):\n", pivot_units, "\n")

# 8) HighValueOrder flag (>5000)
df['HighValueOrder'] = df['Revenue'] > 5000
print("Data with HighValueOrder column:\n", df[['OrderID','Region','Product','Units','UnitPrice','Revenue','HighValueOrder']], "\n")

# 9) Export cleaned file
df.to_csv('sales_cleaned.csv', index=False)
print("Cleaned file exported as sales_cleaned.csv")
```

```python
# Step 1: Import pandas
import pandas as pd

# Step 2: Simple data create karo (file banane ki zarurat nahi)
data = {
    "ID": [1, 2, 3, 4, 5, 6, 7],
    "Name": ["Ali", "Sara", "Rehan", "Zainab", "Ahmed", "Hira", "Bilal"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Sales", "IT"],
    "Salary": [55000, 80000, 60000, 90000, 50000, 65000, 72000],
    "City": ["Lahore", "Karachi", "Islamabad", "Lahore", "Karachi", "Islamabad", "Lahore"]
}

# Step 3: DataFrame banao
df = pd.DataFrame(data)

# Step 4: Data show karo
print("---- Full Data ----")
print(df)

# Step 5: Rows aur Columns count
print("\nTotal Rows aur Columns:", df.shape)

# Step 6: Sirf Name aur Salary columns
print("\nName aur Salary Columns:")
print(df[["Name", "Salary"]])

# Step 7: Lahore wale employees
print("\nLahore wale employees:")
print(df[df["City"] == "Lahore"])

# Step 8: IT Department wale employees
print("\nIT Department employees:")
print(df[df["Department"] == "IT"])

# Step 9: Highest Salary employee
print("\nHighest Salary employee:")
print(df[df["Salary"] == df["Salary"].max()])

# Step 10: Average Salary
print("\nAverage Salary:", df["Salary"].mean())

# Step 11: Department wise average salary
print("\nDepartment wise Average Salary:")
print(df.groupby("Department")["Salary"].mean())

# Step 12: Salary descending order mein sort karo
print("\nSalary Descending Order:")
print(df.sort_values("Salary", ascending=False))

# Step 13: Bonus column (10% of Salary)
df["Bonus"] = df["Salary"] * 0.10
print("\nData with Bonus column:")
print(df)

# Step 14: TotalIncome column (Salary + Bonus)
df["TotalIncome"] = df["Salary"] + df["Bonus"]
print("\nData with TotalIncome:")
print(df)

# Step 15: Top 3 salary wale employees
print("\nTop 3 Salary Employees:")
print(df.sort_values("Salary", ascending=False).head(3))
```
